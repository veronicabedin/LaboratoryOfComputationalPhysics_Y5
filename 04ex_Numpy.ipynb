  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import math 
#construct bool array 
N=100
prime= np.arange(0, N)
print(prime)
mask=np.ones((100), dtype=bool)
print(mask)
mask[0]=mask[1]=0
for i in range(2,N):
    for count in range(2,math.floor(i/2)+1):
        if(i%count==0):
            #print(count, prime[i], " Diversi: ", prime[i]!=count, "    Module:", prime[i]%count!=0)
            mask[i]=0
            break
print(mask)
prime_filt=prime[mask]
print(prime_filt)
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
import numpy.random as npr
npr.seed(123) # fix seed for reproducible result

# create a 2D array of size walkers x steps with values -1 or 1
steps=npr.randint(-1,1,size=(1000,200))
steps[steps == 0] = 1

#Build the actual walking distances for each walker
sum_row=np.zeros((1000,200))
sum_row[:,0]=steps[:,0]

for j in range(1000):
    for i in range(1,200):
        sum_row[j][i]=steps[j][i]+sum_row[j][i-1]

#compute the square alementwise, the mean along columns and the sqrt
sum_row=np.square(sum_row)
sum_row=np.mean(sum_row, axis=0)
sum_row=np.sqrt(sum_row)
plt.xlabel("Time step")
plt.ylabel("Average distance")
plt.plot(sum_row, color ="green")
plt.show()
````

````python
#load the data 
population=np.loadtxt("populations.txt")

#create arrays of the columns
year, hares, lynxes, carrots = population[:, 0], population[:, 1], population[:, 2], population[:, 3]

#plot the 3 populations over the years
plt.xlabel("Year")
plt.ylabel("Population")
plt.plot(year, hares, color ="brown", label='Hares')
plt.plot(year, lynxes, color ="grey", label='Lynxes')
plt.plot(year, carrots, color ="orange", label='Carrots')
plt.legend(loc="upper right")
plt.show()

#compute mean, std, correlations
h_mean, l_mean, c_mean=np.mean(hares), np.mean(lynxes), np.mean(carrots)
h_std, l_std, c_std=np.std(hares), np.std(lynxes), np.std(carrots)
h_l_corr, c_l_corr, h_c_corr=np.corrcoef(hares,lynxes), np.corrcoef(carrots,lynxes), np.corrcoef(hares,carrots)
print("Hares mean and std:", h_mean, "+/-", h_std)
print("Lynxes mean and std:", l_mean, "+/-", l_std)
print("Carrots mean and std:", c_mean, "+/-", c_std)
print("Lynxes-hares correlation coefficient:", h_l_corr[0][1])
print("Carrots-hares correlation coefficient:", h_c_corr[0][1])
print("Lynxes-carrots correlation coefficient:", c_l_corr[0][1])

#highest population each year
max_p=0
highest=[]
for i in range(len(year)):
    if lynxes[i]>carrots[i]:
        max_p=lynxes[i]
        s='lynxes'
    else:
        max_p=carrots[i]
        s='carrots'
    if hares[i]>max_p:
        s='hares'
    max_p=0
    highest.append(s)
print(highest)
````
